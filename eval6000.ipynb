```python
# try to combine history and measurements, e.g. for CPU vs. rq/s

exr.pods['run_seconds'] = exr.pods['run_time'].dt.total_seconds()

run_starts = exr.stats_history.groupby(exr.RUN_VARS)['timestamp'].transform("min")
exr.stats_history['run_seconds'] = exr.stats_history['timestamp'] - run_starts
exr.stats_history
```

```python
sns.relplot(
    kind="line",
    data=exr.stats_history,
    x="run_seconds",
    y="rq_s",
    row="exp_branch",
    col="exp_workload",
)
```

```python
merged = pd.merge(
    left=exr.pods,
    right=exr.stats_history,
     on="run_seconds"
    )

merged
```

```python
mergemelt = merged.melt(
    id_vars=['exp_workload_x', 'exp_branch_x', "run_seconds"],
    value_vars=['cpu_usage', 'rq_s']
)

mergemelt
```

```python
sns.relplot(
    kind="line",
    data=mergemelt,
    x="run_seconds",
    y="value",
    hue="variable",
    row="exp_branch_x",
    col="exp_workload_x",
)
```
